```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load the fer2013 dataset using pandas
data = pd.read_csv('Data Set/fer2013.csv')

# Extract 36 grayscale images from the dataset and resize them
images = []
for i in range(36):
    image = np.fromstring(data['pixels'][i], dtype=int, sep=' ').reshape((48, 48))
    image = cv2.resize(image.astype('uint8'), (48, 48))
    images.append(image)

# Normalize the pixel values to be between 0 and 1
images = np.array(images) / 255.0

# Create a 6x6 matrix and populate it with the grayscale images
matrix = np.zeros((6*48, 6*48))
for i in range(6):
    for j in range(6):
        matrix[i*48:(i+1)*48, j*48:(j+1)*48] = images[i*6+j]

# Display the matrix
plt.imshow(matrix, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import os
import shutil
import matplotlib.pyplot as plt
from PIL import Image

# Define the path to the folder containing the CK+ dataset
dataset_path = 'CK+ DataSet/CK+48'

# Define a dictionary that maps folder names to emotion labels
folder_labels = {'anger': 'anger', 'contempt': 'contempt', 'disgust': 'disgust', 'fear': 'fear', 'happy': 'happy', 'sadness': 'sadness', 'surprise': 'surprise'}

# Define a list of emotion labels to display in the matrix
emotion_labels = ['anger', 'contempt', 'disgust', 'fear', 'happy', 'sadness', 'surprise', 'neutral']

# Create a 2x4 grid for displaying the images
fig, axs = plt.subplots(2, 4, figsize=(10, 6))

# Loop through each emotion label in the list and select one image from the corresponding folder
for i, emotion_label in enumerate(emotion_labels):
    folder_name = emotion_label
    folder_path = os.path.join(dataset_path, folder_name)
    # Check if the folder is a valid folder containing images
    if os.path.isdir(folder_path) and len(os.listdir(folder_path)) > 0:
        # Select the first image in the folder
        image_name = os.listdir(folder_path)[0]
        image_path = os.path.join(folder_path, image_name)
        # Rename the image file based on the folder's emotion label
        new_image_name = f"{emotion_label}_{image_name}"
        new_image_path = os.path.join(folder_path, new_image_name)
        shutil.move(image_path, new_image_path)
        # Read the image file and display it in the corresponding subplot of the matrix
        image = Image.open(new_image_path)
        axs[i//4, i%4].imshow(image, cmap='gray')
        axs[i//4, i%4].set_title(emotion_label)
        axs[i//4, i%4].axis('off')
        axs[i//4, i%4].set_xlim([0, image.size[0]])
        axs[i//4, i%4].set_ylim([image.size[1], 0])

plt.show()
```
